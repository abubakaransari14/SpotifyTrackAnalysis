```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading
data = r'C:\Users\uzair\Downloads\Spotify\dataset.csv'
pd.set_option("display.max_columns", None)
df = pd.read_csv(data)
df.head() #114000 rows × 21 columns
```

```python
# data exploration
print('Shape:' , df.shape)
print('Columns: ', df.columns)
print('Duplicated Track: ',df.duplicated(subset=['track_id']).sum())
print('x-----------------------------------------------------------------------------------------------x')
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Data Cleaning
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
df = df.drop_duplicates(subset=['track_id'])
df = df.dropna(subset=['popularity','danceability','energy','valence'])
dfClean = df.copy()
```

```python
hitThreshold = dfClean['popularity'].quantile(0.80)
dfClean['isHit'] = dfClean['popularity'] >= hitThreshold
dfClean[['track_name','artists','popularity','isHit']].head()
```

```python
audioFeatures = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'liveness',
    'speechiness'
]
hitDNA = dfClean.groupby('isHit')[audioFeatures].mean()

hitDNA
```

```python
featureCorr = dfClean[audioFeatures + ['popularity']].corr()

featureCorr['popularity'].sort_values(ascending=False)
```

```python
artistStats = dfClean.groupby('artists').agg(
    trackCount=('track_id','count'),
    avgPopularity=('popularity','mean'),
    avgDanceability=('danceability','mean'),
    avgEnergy=('energy','mean'),
    avgValence=('valence','mean')
).reset_index()

artistStats.head()
```

```python
artistHitRate = dfClean.groupby('artists')['isHit'].mean().reset_index()

artistHitRate.rename(columns={'isHit':'hitRate'}, inplace=True)

artistStats = artistStats.merge(artistHitRate, on='artists')
```

```python
artistStats
```

```python
emergingArtists = artistStats[
    (artistStats['trackCount'] >= 3) &
    (artistStats['trackCount'] <= 10)
]

emergingArtists = emergingArtists.sort_values(
    by=['avgPopularity','hitRate'],
    ascending=False
)

emergingArtists.head(20)
```

```python
genreStats = dfClean.groupby('track_genre').agg(
    avgPopularity=('popularity','mean'),
    volume=('track_id','count')
).reset_index()

genreStats.head()
```

```python
totalTracks = genreStats['volume'].sum()

genreStats['marketShare'] = genreStats['volume'] / totalTracks

genreStats.sort_values('avgPopularity', ascending=False).head(10)
```

```python
genreHitRate = dfClean.groupby('track_genre')['isHit'].mean().reset_index()

genreHitRate.rename(columns={'isHit':'hitRate'}, inplace=True)

genreHitRate.sort_values('hitRate', ascending=False).head(10)
```

```python
marketOpportunity = genreStats.sort_values(
    ['avgPopularity','volume'],
    ascending=[False,True]
)

marketOpportunity.head(10)
```

```python
correlation = artistStats['trackCount'].corr(artistStats['avgPopularity'])

print("Correlation between productivity and popularity:", correlation)
```
